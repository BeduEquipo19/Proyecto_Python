### Paso a: Identificación del problema
Problema: **Alta percepción de inseguridad en el estado de Quintana Roo**

Planteamiento:

La percepción de inseguridad, es decir, el sentimiento que tiene una persona de ser victima de un delito independientemente de la probabilidad de serlo, es un problema que vulnera la calidad de vida de las personas, las cuales dejan de realizar sus actividades cotidianas, lo que deriva en la limitación del esparcimiento social, interfiere en el proceso de cohesión social, y en algunos casos puede generar problemas mayores.  
  
Justificación del problema: El estado de Quintana Roo tiene múltiples atractivos turísticos con un impacto económico significativo. Sin embargo, los residentes del estado están sujetos a intensas dinámicas sociales derivadas de la actividad turística, mismas que no siempre son medidas o sondeadas, principalmente en el aspecto de seguridad. Por tal motivo, el gobierno del estado de Quintana Roo desea conocer cuáles son los principales elementos de inseguridad que perciben los ciudadanos del estado en sus entornos de vivienda, trabajo, transporte y socialización. Cabe resaltar que de todos los centros urbanos del país, la zona metropolitana de Cancún es la que tiene el segundo mayor índice de inseguridad.
